### Urdu to roman Function 

```python
def urdu2roman(word):
    #looking at the string character by character and based on character position in word returning english couterpart
    rw=''
    for char in word:
        #in case if the character is an digit return as it is
        if char.isdigit():    
            rw = rw + char
            
        #in case if the character is an English alphabet return as it is
        elif (ord(charr) >= 65) and (ord(char) <= 90):    
            rw = rw + char
            
        #in case if the character is an English alphabet return as it is
        elif (ord(charr) >= 97) and (ord(char) <= 122):
            rw = rw + char
            
        # in case the character is from urdu                             
        elif word.startswith(char):
            #print(urdu_dict[ord(char)]['start'])
            rw = rw + (urdu_to_roman_dict[hex(ord(char))]['start'])
        elif word.endswith(char):
            #print(urdu_dict[ord(char)]['end'])
            rw = rw + (urdu_to_roman_dict[hex(ord(char))]['end'])
        else:
            #print(urdu_dict[ord(char)]['mid'])
            rw = rw + (urdu_to_roman_dict[hex(ord(char))]['mid'])
    return(rw)
```

### Roman2Urdu Function implimentation 

```python
def roman2urdu(word1):
    word=word1
    uw=''

# BASIC SINGLE-LETTER RULES 
    
    #Defining Some Rules
    if word.isdigit():    
        return word
    if word == 'B':
        word = word.replace('B','BY')     
    if word == 'K':
        word = word.replace('K','KE')
    if word == 'WO':
        word = word.replace('WO','OH')
    if word == 'C':
        word = word.replace('C','SY')
        
    if word.startswith("AA"):
        word = word.replace('AA',roman_to_urdu_dict['AA'],1)


# BASIC STARTING WORD LATTER RULES


    #If the word Starts with a vovel replace it with A => alif    
    if word.startswith("E"):
        word = word.replace('E','AE',1)
    if word.startswith("AR"):
        word = word.replace('AR','AR',1)
    if word.startswith("I"):
        word = word.replace('I','A',1)
    if word.startswith("O"):
        word = word.replace('O','AO',1)
    if word.startswith("U"):
        word = word.replace('U','A',1)
        
    if word.startswith("OI"):
        word = word.replace('OI','AY',1)
                

# CHANGING CERTAIN LETTERS INTO VOWEL FROM RULES
    
    if "W" in word:
        word = word.replace('W','O')
    if "V" in word:
        word = word.replace('V','O')
    if "AR" in word and word.startswith("AR")==False:
        word = word.replace('AR','R')
    if "AE" in word:
        word = word.replace('AE',roman_to_urdu_dict['AE'])
    if "SH" in word:
        word = word.replace('SH',roman_to_urdu_dict['SH'])
    if "CH" in word:
        word = word.replace('CH',roman_to_urdu_dict['CH'])
    if "KH" in word:
        word = word.replace('KH',roman_to_urdu_dict['KH'])
    if "GH" in word:
        word = word.replace('GH',roman_to_urdu_dict['GH'])

# VOWEL TO VOWLE CHANGING RULES
    
    #Changes vowel sequence     
    if "AA" in word:
        word = word.replace('AA','A')
    if "AI" in word:
        word = word.replace('AI','Y')
    if "EI" in word:
        word = word.replace('EI','Y')
    if "EE" in word:
        word = word.replace('EE','Y')
    if "IE" in word:
        word = word.replace('IE','Y')
    if "OO" in word:
        word = word.replace('OO','O')
    if "AU" in word:
        word = word.replace('AU','O')
    if "OU" in word:
        word = word.replace('OU','O')
    if "U" in word:
        word = word.replace('U','O')

# CHANGING SINGLE LETTTER VOWEL TO ADVANCE-MORE-LETTER VOWEL
    
    if word.endswith("I"):
        word = word.replace('I','Y') 
    if word.endswith("AIN"):
        word = word.replace('AIN','EIN') 
    if "I" in word:
        word = word.replace('I','Y')
    if not(word.endswith("E")):
        word = word.replace('E','Y',1)
    if "C" in word:
        word = word.replace('C','K')
    if word.endswith("EY"):
        word = word.replace('EY','Y') 
    if "X" in word:
        word = word.replace('X','KS')
        
        
        
    #MAPPING fROM dictionary    
    for char in word:
        if char in roman_to_urdu_dict:
            word = word.replace(char,roman_to_urdu_dict[char])    
    
    return word
    
```

### Removing consucative duplicates 

```python
# Removing consucative duplicates
def rem_consec_dup(word):
    new_w = ""
    prev = ""
    for c in word:
        if len(new_w) == 0:
            new_w += c
            prev = c
        if c == prev:
            continue
        else:
            new_w += c
            prev = c
    return new_w
```
